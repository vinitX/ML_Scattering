```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f=open("LXCat0.txt","w")
f.write('----------------------\n')
f.write('ELASTIC\nHard\n')
f.write('1.371541e-4 / mass ratio\n')
f.write('----------------------\n')
n=500
E=np.logspace(-3,np.log10(3),n)
sigma_m=5e-20
for i in range(n):
    f.write(str(E[i])+' '+str(sigma_m)+'\n')
#f.write('----------------------\n')
```

```python
f.write('----------------------\n')
f.write('EFFECTIVE\nMaxwell\n')
f.write('1.371541e-4 / mass ratio\n')
f.write('----------------------\n')
n=500
E=np.logspace(-3,np.log10(3),n)
sigma_m=5e-20*E**-0.5
for i in range(n):
    f.write(str(E[i])+' '+str(sigma_m[i])+'\n')
#f.write('----------------------\n')

plt.loglog(E,(sigma_m)*1e20)
plt.show()
```

```python
n=1000
E=np.logspace(-3,1,n)
f.write('----------------------\n')
f.write('ELASTIC\nPower\n')
f.write('3.4288525e-05 / mass ratio\n')
f.write('----------------------\n')
sigma_m=10e-20*E**-0.5
for i in range(n):
    f.write(str(E[i])+' '+str(sigma_m[i])+'\n')
    
f.write('----------------------\n')
f.write('ATTACHMENT\nPower\n')
f.write('3.4288525e-05 / mass ratio\n')
f.write('----------------------\n')
sigma_loss=1*1e-20*E**0.5
for i in range(n):
    f.write(str(E[i])+' '+str(sigma_loss[i])+'\n')

plt.loglog(E,(sigma_m+sigma_loss)*1e20)
plt.show()
```

```python
n=1000
E=np.logspace(-3,np.log10(3),n)

f.write('----------------------\n')
f.write('ELASTIC\nReid\n')
f.write('1.371541e-4 / mass ratio\n')
f.write('----------------------\n')
sigma_m=6e-20
for i in range(n):
    f.write(str(E[i])+' '+str(sigma_m)+'\n')


f.write('----------------------\n')
f.write('EXCITATION\nReid\n')
f.write('0.2 / Thresold Energy\n')
f.write('1.371541e-4 / mass ratio\n')
f.write('----------------------\n')
sigma_exc=10e-20*(E-0.2)*np.heaviside(E-0.2,1)
for i in range(n):
    f.write(str(E[i])+' '+str(sigma_exc[i])+'\n')
    

plt.loglog(E,(sigma_m+sigma_exc)*1e20)
plt.show()
```
